```python
import numpy as np
import pandas as pd
```

```python
roll_no = [i for i in range(1,11)]
roll_no
```

```python
names = ['Abhishek Jha', 'Shubham Rajput', 'Yash Patel', 'Scarlett Johansson', 'Shivam Sarang', 
         'Yash Shah', 'Emma Watson', 'Alexandra Daddario', 'Dhairya Patel', 'Emilia Clarke']
names
```

```python
places = ['Dindoli', 'Bhatar', 'Vesu', 'Dumas', 'Rander', 
          'Adajan', 'Vesu', 'City Light', 'Dumas', 'Dindoli']
places
```

```python
df = pd.DataFrame({'Roll_No': roll_no, 'Names': names, 'Places': places})
df
```

```python
gender = ['Male', 'Male', 'Male', 'Female', 'Male', 'Male', 'Female', 'Female', 'Male', 'Female']
df['Gender'] = gender
df
```

```python
marks_pds = list(np.random.randint(30, 101, 10))
marks_cc = list(np.random.randint(30, 101, 10))
df['Marks_PDS'] = marks_pds
df['Marks_CC'] = marks_cc
df
```

```python
df = df.set_index('Roll_No')
df
```

```python
print(df.to_string(index = False))
```

```python
print('Statistical Summary: \n')
print(df.describe())

print('\nCorrelation: \n')
print(df.corr())
```

```python
print(df.groupby('Places').count())
```

```python
print(df.groupby('Places')['Gender'].value_counts())
```

```python
df.groupby('Gender').get_group('Male')
```

```python
df.groupby('Gender').agg([np.sum, np.mean, np.median, np.std])['Marks_PDS']
```

```python
df.groupby('Gender').agg([np.sum, np.mean, np.median, np.std])['Marks_CC']
```

```python
df.groupby('Gender').agg([np.size])
```

```python
score = lambda x: ((x - x.mean()) / x.std()) * 10
(df.groupby(['Gender'])['Marks_PDS', 'Marks_CC']).transform(score)
```

```python
df.groupby('Places').filter(lambda x: len(x) >= 2)
```
