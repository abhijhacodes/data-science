```python
from bs4 import BeautifulSoup, Comment
import requests, csv
import re
import urllib3
import pandas as pd
from urllib.request import Request, urlopen
```

```python
# Contents of URL

url = 'https://en.wikipedia.org/wiki/Comparison_of_text_editors'
res = requests.get(url)
html_page = res.content
soup = BeautifulSoup(html_page, 'html.parser')
contents = soup.find_all(text=True)
print('Contents of the URL:')
contents
```

```python
# Title of HTML page

title = soup.title.get_text()
print('Title of the HTML page:')
title
```

```python
# Reading paragraph tags

total = len(soup.find_all('p'))
print(f'Total paragraph tags = {total}')

print('Paragraph tags on this page are:\n')
for paragraph in soup.find_all('p'):
    print(paragraph.get_text())
    
```

```python
# Reading anchor tags

total = len(soup.find_all('a'))
print(f'Total anchor tags = {total}')

print('Anchor tags on this page are:')
for anchor in soup.find_all('a'):
    print(anchor.get_text())
    
```

```python
# Reading class tags

total = len(soup.find_all('class'))
print(f'Total class tags = {total}')

print('Class tags on this page are:')
for clas in soup.find_all('class'):
    print(clas.get_text())
    
```

```python
# Reading image tags

total = len(soup.find_all('img'))
print(f'Total image tags = {total}')

print('Image tags on this page are:\n')
for image in soup.find_all('img'):
    print(f'{image}\n')
    
```

```python
# Reading comment tags

total = len(soup.find_all(string=lambda text: isinstance(text, Comment)))
print(f'Total comment tags = {total}')

print('Comment tags on this page are:\n')
for comment in soup.find_all(string=lambda text: isinstance(text, Comment)):
    print(comment)
    
```

```python
for href in soup.find_all('a', attrs={'href': re.compile("^http://")}):
    print(f'{href}\n')
```

```python
rank_url = 'https://socialblade.com/youtube/top/50'
headers = {'User-agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_13_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/70.0.3538.77 Safari/537.36'}
           
try:
    source = Request(rank_url, headers=headers)
    page = urlopen(source)
    file = open('topyoutubers.csv', 'w', encoding='utf-8')
    writer = csv.writer(file, lineterminator='\n')
    writer.writerow(['Rank', 'Channel Name', 'Uploads', 'Subscribers', 'Views'])
    soup = BeautifulSoup(page, 'html.parser')
    rows = soup.find('div', attrs={'style': 'float: right; width: 900px;'}).find_all('div', recursive=False)[4::]
    for row in rows:
        rank = row.find('div', attrs={'style': 'float: left; width: 50px; color:#888;'}).text.strip()
        channel = row.find('a').text.strip()
        temps = row.find_all('span', attrs={'style': 'color:#555;'})
        uploads = temps[0].text.strip()
        subscribers = row.find('div', attrs={'style': 'float: left; width: 150px;'}).text.replace('\n', '').strip()
        views = temps[1].text.strip()
        writer.writerow([rank, channel, uploads, subscribers, views])
    file.close()
           
except Exception as e:
     print(e)
```

```python
url = "https://boston.craigslist.org/search/npo" 
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_13_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/70.0.3538.77 Safari/537.36'} 

try: 
    source = Request(url, headers=headers) 
    page = urlopen(source) 
    file = open('jobopenings.csv', 'w', encoding='utf-8') 
    writer = csv.writer(file, lineterminator ="\n") 
    writer.writerow(['Job', 'Location']) 
    soup = BeautifulSoup(page, 'html.parser')
    rows = soup.find('ul', class_="rows").find_all('div', class_="result-info")
    for row in rows:
        job = row.find('h3', class_="result-heading")
        loc = row.find('span', class_="result-meta").find('span', class_="result-hood")
        if loc is None or job is None:
            job = row.find('h3', class_="result-heading").text.strip()
            loc = row.find('span', class_="result-meta").find('span',
            class_="nearby").text.strip()
            writer.writerow([job, loc])
        else:
            job, loc = job.text.strip(), loc.text.strip()
            writer.writerow([job, loc])
    file.close()
    
except Exception as e:
    print(e)
```

```python
url = "https://en.wikipedia.org/wiki/Comparison_of_text_editors"
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_13_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/70.0.3538.77 Safari/537.36'}

try:
    source = Request(url, headers=headers)
    page = urlopen(source)
    file = open('wikipedia.csv', 'w', encoding='utf-8')
    writer = csv.writer(file, lineterminator ="\n")
    writer.writerow([' ', 'Microsoft Windows', 'macOS', 'Linux', 'BSD', 'UNIX', 'OpenVMS'])
    soup = BeautifulSoup(page, 'html.parser')
    rows = soup.find('table', style="max-width: 30em; table-layout: fixed").find('tbody').find_all('tr')[1::]
    for row in rows:
        lst = []
        text_editor = row.find('th').text.split()[0]
        for a in row.find_all('td'):
            if a.img is None:
                lst.append("Partial")
            else:
                lst.append(a.img['alt'])
        win, mac, lin, bsd, uni, vms = lst[0], lst[1], lst[2], lst[3], lst[4], lst[5]
        writer.writerow([text_editor, win, mac, lin, bsd, uni, vms])
    file.close()
    
except Exception as e:
    print(e)
```

```python
url = "https://www.flipkart.com/search?q=iphone&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off" 
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_13_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/70.0.3538.77 Safari/537.36'} 

try: 
    source = Request(url, headers=headers) 
    page = urlopen(source) 
    file = open('FlipKart.csv', 'w', encoding='utf-8') 
    writer = csv.writer(file, lineterminator ="\n") 
    writer.writerow(['Iphone Model', 'Rating', 'ROM', 'Display', 'Front Camera', 'Processor', 'Price']) 
    soup = BeautifulSoup(page, 'html.parser') 
    rows = soup.find('div', style="flex-grow:1;overflow:auto").find_all('div', class_="_1AtVbE col-12-12")[0:-2] 
    
    for row in rows: 
        iphone = row.find('div', class_="_4rR01T", recursive=True) 
        star = row.find('div', class_="_3LWZlK", recursive=True) 
        price = row.find('div', class_="_30jeq3 _1_WHN1", recursive=True)
        specs = row.find_all('li', class_="rgWa7D")
        cnt, temp = 0, []
        for s in specs:
            temp.append(str(*s))
            cnt += 1
            if cnt == 4:
                break
        price = row.find('div', class_="_30jeq3 _1_WHN1", recursive=True)
        if iphone is None:
            iphone = "Model Not Available"
        else:
            iphone = iphone.text.strip()
        if star is None:
            star = "Unrated"
        else:
            star = star.text.strip() + "/5"
        if price is None:
            price = "Not Known"
        else:
            price = "Rs. " + price.text.strip().replace('₹', '')
        rom, display, frnt, procc = temp[0], temp[1], temp[2], temp[3]
        writer.writerow([iphone, star, rom, display, frnt, procc, price])
    file.close()
        
except Exception as e:
    print(e)
```
