```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sales_data = pd.read_csv('company_sales_data.csv')
sales_data
```

```python
month_number = sales_data.loc[:, ['month_number']]
total_profit = sales_data.loc[:, ['total_profit']]

plt.plot(month_number, total_profit, color='red',
         linestyle='dotted', linewidth=3,
         label='Month-wise Profit data of last year',
         marker = 'o', markerfacecolor = 'red')

plt.legend(loc='lower right')
plt.xlabel('Month Number')
plt.xticks(list(sales_data['month_number']))

plt.ylabel('Profit in dollar')
plt.yticks([100000, 200000, 300000, 400000, 500000])

plt.show()
```

```python
sales_data['facecream'].plot(ls='--',lw=1,marker='s',color='b',label='FaceCream Sales Data')
# plt.plot(sales_data['facecream'], ls='--', lw=1, marker='s', color='b', label='FaceCream Sales Data')

sales_data['facewash'].plot(ls='-',lw=1.1,marker='+',color='g',label='FaceWash Sales Data')

sales_data['toothpaste'].plot(ls='-.',lw=1.2,marker='o',color='r',label='ToothPaste Sales Data')

sales_data['bathingsoap'].plot(ls=':',lw=1.3,marker='*',color='c',label='BathingSoap Sales Data')

sales_data['shampoo'].plot(ls='--',lw=1.4,marker='p',color='m',label='Shampoo Sales Data')

sales_data['moisturizer'].plot(ls='-',lw=1.5,marker='P',color='y',label='Moisturizer Sales Data')

plt.legend(loc='upper left')
plt.xlabel('Month Number')
plt.xticks(list(sales_data['month_number']))

plt.ylabel('Sales unit in numbers')
plt.yticks([1000, 2000, 4000, 6000, 8000, 10000, 12000, 15000, 18000])

plt.show()
```

```python
month_number = list(sales_data['month_number'])
toothpaste = list(sales_data['toothpaste'])

plt.scatter(month_number, toothpaste, label='Tooth paste Sales data')
plt.legend(loc='upper left')
plt.title('Tooth paste Sales data')
plt.grid(ls='--', lw=1)

plt.xlabel('Month Number')
plt.xticks(month_number)

plt.ylabel('Number of units sold')
plt.yticks([4500, 5000, 5500, 6000, 6500, 7000, 7500, 8000])

plt.show()
```

```python
facecream = list(sales_data['facecream'])
facewash = list(sales_data['facewash'])
month_number = list(sales_data['month_number'])

plt.bar([month-0.25 for month in month_number], facecream, color='orange',
        width=0.25, label='Face cream sales data', align='edge')

plt.bar([month for month in month_number], facewash, color='green',
        width=0.25, label='Face wash sales data', align='edge')

plt.grid(ls='--', lw=1)
plt.legend(loc='upper left')
plt.title('Facewash and Facecream Sales data')

plt.xlabel('Month Number')
plt.xticks(month_number)

plt.ylabel('Number of units sold')

plt.show()
```

```python
total_profit = list(sales_data['total_profit'])
profit_range = [150000, 175000, 200000, 225000, 250000, 300000, 350000]

plt.hist(total_profit, profit_range)
plt.title('Profit data')

plt.xlabel('Profit range in dollar')
plt.xticks(profit_range)

plt.ylabel('Actual profit in dollar')
plt.show()
```

```python
s1 = sales_data['facecream'].sum()
s2 = sales_data['facewash'].sum()
s3 = sales_data['toothpaste'].sum()
s4 = sales_data['bathingsoap'].sum()
s5 = sales_data['shampoo'].sum()
s6 = sales_data['moisturizer'].sum()

sales = [s1, s2, s3, s4, s5, s6]
labels = ['FaceCream', 'FaseWash', 'ToothPaste', 'Bathing soap', 'Shampoo', 'Moisturizer']
colors = ['lime', 'magenta', 'orange', 'palegreen', 'cornflowerblue', 'yellow']
plt.pie(sales, labels=labels, colors=colors, autopct='%1.1f%%')

plt.title('Sales data')
plt.legend(bbox_to_anchor=(1.35,0), loc='lower right')
plt.axis('equal')
plt.show()
```

```python
facewash = list(sales_data['facewash'])
bathingsoap = list(sales_data['bathingsoap'])

fig,axes = plt.subplots(1,2,figsize=(10,4))

axes[0].plot(bathingsoap, marker='s', markerfacecolor='black', color='lime') 
axes[0].set_title('Sales data of a Bathing Soap') 

axes[1].plot(facewash, marker='^', markerfacecolor='red', color='royalblue') 
axes[1].set_title('Sales data of a Facewash')

for ax in axes: 
    ax.set_xlabel('Month Number') 
    ax.set_xticks(month_number) 
    ax.set_ylabel('Sales units in number')
    
plt.show()
```

```python
facecream = sales_data['facecream']
facewash = sales_data['facewash']
toothpaste = sales_data['toothpaste']
bathingsoap = sales_data['bathingsoap']
shampoo = sales_data['shampoo']
moisturizer = sales_data['moisturizer']

labels = ['FaceCream', 'FaseWash', 'ToothPaste', 'Bathing soap', 'Shampoo', 'Moisturizer']
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'magenta', 'green']

plt.stackplot(month_number, facecream, facewash, toothpaste, bathingsoap, shampoo, moisturizer, 
              labels=labels, colors=colors, lw=5)

plt.legend(loc='upper left')
plt.title('All products sales data using stack plot')
plt.xlabel('Month number')
plt.ylabel('Sales units in number')
# plt.axis('equal')

plt.show()
```

```python
import datetime as DT
from matplotlib.dates import date2num

data = [(DT.datetime.strptime('2016-10-03', "%Y-%m-%d"), 772.559998),
        (DT.datetime.strptime('2016-10-04', "%Y-%m-%d"), 776.429993),
        (DT.datetime.strptime('2016-10-05', "%Y-%m-%d"), 776.469971),
        (DT.datetime.strptime('2016-10-06', "%Y-%m-%d"), 776.859985),
        (DT.datetime.strptime('2016-10-07', "%Y-%m-%d"), 775.080017)]

dates = [date2num(date) for (date, value) in data]
closing = [value for (date, value) in data]

fig = plt.figure()
graph = fig.add_subplot(111)
graph.plot(dates, closing, 'b-o')
graph.set_xticks(dates)
graph.set_xticklabels([date.strftime("%Y-%m-%d") for (date, value) in data])

plt.grid(which='major', ls='-', lw=0.8, color='red')
plt.minorticks_on()
plt.grid(which='minor', ls=':', lw=0.7, color='black')

plt.xlabel('Dates')
plt.ylabel('Closing values')
plt.title('Closing stocks of company')
plt.tick_params(which='both', top=True, left=False, right=True, bottom=False)
plt.show()
```
